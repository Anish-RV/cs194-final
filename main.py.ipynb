```python
def get_3d_pts(pts):
    three_d_pts = []
    for pt in pts:
        x = int(input("X: "))
        y = int(input("Y: "))
        z = int(input("Z: "))
        three_d_pts.append([x,y,z])
    return three_d_pts
three_d_pts = get_3d_pts(pts)
```

```python
# Initialize bboxes
def init_bboxes(pts, size):
    bbox_lst = []
    for pt in pts:
        bbox = [pt[0]-(size/2), pt[1]-(size/2), size, size]
        bbox_lst.append(bbox)
    return bbox_lst
bboxes = init_bboxes(pts, 8)

# Initialize trackers
def init_trackers(frame, bboxes):
    tracker_lst = []
    for bbox in bboxes:
        tracker = cv.TrackerCSRT_create()
        ok = tracker.init(frame, tuple(bbox))
        if not ok:
            print("Error in setting up tracker")
            return None
        tracker_lst.append(tracker)
    return tracker_lst

tracker_lst = init_trackers(video[0], bboxes)
```

```python
def update_trackers(tracker_lst, video):
    calc_pts = []
    for i in range(0,len(video)):
        frame = video[i]
        calc_pts.append([])
        for tracker in tracker_lst:
            ok, bbox = tracker.update(frame)
            if ok:
                calc_pts[i].append([int(bbox[0]+bbox[2]/2), int(bbox[1]+bbox[3]/2)])
    return calc_pts
calc_pts = update_trackers(tracker_lst, video)
```

```python
def display_pts(calc_pts, video):
    for i in range(len(video)):
        frame = video[i]
        for pt in calc_pts[i]:
            cv.circle(frame, tuple(pt), 4, (0,255,0), -1)
        cv.imshow("Tracking", frame)
        k = cv.waitKey(30)
    return
display_pts(calc_pts, video)
```

```python
sk_vid.vwrite("box_tracking.mp4", video, outputdict={'-pix_fmt':'yuv420p'})
```

```python
def calibration(calc_pts, three_d_pts, video):
    matrices = []
    three_d_pts = np.array(three_d_pts)
    world_pts = np.hstack((three_d_pts, np.ones((len(three_d_pts), 1))))
    for i in range(len(video)):
        i_pts = calc_pts[i]
        matrix = np.array(np.linalg.lstsq(world_pts, i_pts))[0]
        matrices.append(matrix.T)
    return matrices

def draw(img, imgpts):
    imgpts = np.int32(imgpts).reshape(-1,2)
    # draw ground floor in green
    img = cv.drawContours(img, [imgpts[:4]],-1,(0,255,0),-3)
    # draw pillars in blue color
    for i,j in zip(range(4),range(4,8)):
        img = cv.line(img, tuple(imgpts[i]), tuple(imgpts[j]),(255),3)
    # draw top layer in red color
    img = cv.drawContours(img, [imgpts[4:]],-1,(0,0,255),3)
    return img

def cube(video, axes, matrices):
    cube_vid = []
    for i in range(len(video)):
        axes_pts = np.dot(matrices[i], axes).T
        drawn = draw(video[i], axes_pts)
        cv.imshow("Tracking", drawn)
        k = cv.waitKey(30)
        cube_vid.append(drawn)
    return cube_vid

matrices = calibration(calc_pts, three_d_pts, video)
axes = np.float32([[2,3,0,1], [2,4,0,1], [3,4,0,1], [3,3,0,1],
                   [2,3,-1,1],[2,4,-1,1],[3,4,-1,1],[3,3,-1,1] ]).T
cube_vid = cube(video, axes, matrices)
```
